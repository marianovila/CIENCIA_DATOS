## DISTRIBUCIÓN NORMAL - GRÁFICO

```python
media=0 #media de distribucion
de=1 #desv estandar
datos=np.random.normal(media,de,1000)

#crea el histograma
#Density True el hist se normaliza de manera que el area total debajo del hist es 1

plt.hist(datos,bins=100,density=True)

plt.title('Distribución normal')
plt.xlabel('Valores')
plt.ylabel('Densidad')

plt.show()
```

## SCATTER - GRÁFICO DE DISPERSIÓN

```python
#scatter, dispersión

x=[1,2,3,4,5]
y=[2,4,6,8,10]

plt.scatter(x,y)

plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico')

plt.show()
```

```python
# Datos para el gráfico de dispersión
estudiantes = ['María', 'José', 'Pedro', 'Miguel', 'Josefina']
matematicas = [80, 75, 90, 65, 85]
quimica = [70, 85, 95, 80, 60]

# Crear el gráfico de dispersión
plt.scatter(matematicas, quimica)

# Personalizar el gráfico
plt.title('Calificaciones de Estudiantes')
plt.xlabel('Matemáticas')
plt.ylabel('Química')

# Etiquetar cada punto con el nombre del estudiante
for i in range(len(estudiantes)):
    plt.annotate(estudiantes[i], (matematicas[i], quimica[i]))

# Mostrar el gráfico
plt.show()
```

## USO DEL PARÁMETRO ALPHA
* El uso del parámetro alpha en un gráfico de dispersión te permite controlar la transparencia de los puntos, lo que puede ser útil para evitar superposiciones, visualizar la densidad de puntos, destacar puntos específicos o comparar grupos de puntos. La elección del valor de alpha dependerá del efecto visual deseado y de las características específicas de tus datos.

```python
n = 1024 #numero de puntos
np.random.seed(32) #semilla aleatoria

x = np.random.normal(0,1,n) 
y = np.random.normal(0,1,n)

#generacion de puntos aleatorios: 
# https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html
# Asigno transparencia con alpha

plt.figure()
plt.scatter(x,y,alpha=0.25)
plt.show()
```

## USO DEL PARÁMETRO CMAP
* En matplotlib, existen diferentes colormaps predefinidos que se pueden utilizar, como "viridis", "jet", "hot", "cool", entre otros. Cada colormap tiene su propio esquema de colores y puede transmitir diferentes significados o características en una representación gráfica.
* El parámetro cmap se utiliza en varias funciones de matplotlib, como scatter, imshow, pcolormesh, entre otras, para especificar el colormap que se utilizará. Por ejemplo, al llamar a scatter con el parámetro cmap, los valores numéricos proporcionados se asignarán a colores según el colormap especificado.

```python
# Para la segunda imagen necesito una tercer columna
# Creo un nuevo array que calcule la distancia de cada punto al origen (0,0)
z = (x**2 + y**2)**0.5

# Dejo a matplotlib que asigne un color por defecto a la columna z
plt.figure()
plt.scatter(x,y,c=z,cmap='Spectral') #cmap='Spectral'   // cmap='bwr_r'
plt.show()
```

## PIE - GRÁFICO DE TORTA

```python
#pie, grafico torta

etiquetas=['manzanas','platanos','naranjas','mangos']
cantidades=[30,45,25,20]

plt.pie(cantidades, labels=etiquetas,autopct='%1.1f%%')

plt.title('Gráfico')

plt.show()
```

```python
#pie torta

etiquetas2=['gatos','perros','pajaros','peces']
cantidades2=[35,45,15,5]

explode=(0.1,0,0,0)

# Colores de cada porción
colores = ['red', 'blue', 'green', 'yellow']

plt.pie(cantidades2, labels=etiquetas2,autopct='%1.1f%%',explode=explode, colors=colores)

plt.title('Ventas por marca')

plt.show()    
            
```
