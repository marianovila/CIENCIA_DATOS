## Debido a la gran cantidad de crmenes y casos sin resolver, la polica local ha decidido implementar su propio sistema de reconocimiento de sospechosos con la tcnica basada en el uso del DNA.   

Para esto la polica mantiene dos listas de informacin; la primera contiene el nombre de las personas registradas en la regin (nombre y apellido), la otra, un cromosoma representativo del DNA de cada una de esas personas.    

Por simplicidad, un cromosoma se considera como una cadena de 0 (ceros) y 1 (unos), de largo 20.

`El mtodo para determinar el sospechoso, es el siguiente:`

* Se obtiene una muestra del cromosoma del autor del delito (20 caracteres)
* Se busca en la lista de cromosomas, aquel cromosoma que es ms parecido a la muestra. El ms parecido se define como el cromosoma que tiene ms genes (caracteres) iguales a la muestra.
* Al terminar la bsqueda, se muestra el nombre de la persona cuyo cromosoma es sospechoso, con el porcentaje de parentesco.

**La informacon inicial del programa se debe ingresar directamente en el cdigo, es decir, nombres y cromosomas, en cambio la secuencia encontrda en la escena del crimen, debe ser ingresada por el usuario.**

Desarrolle un programa que lleve a cabo la bsqueda descrita a partir de una muestra de entrada.

Recuerde que como se trata de dos listas, la informacin del nombre como la de los cromosomas, deben estar en la misma posicin en ambas listas.

Consideremos, personas como Pedro, Juan y Diego. Sus secuencias sern

00000101010101010101
00101010101101110111
00100010010000001001

## SOLUCIN CON LISTAS Y BUCLE

```python
nombres = ['Pedro', 'Juan', 'Diego']
cromosomas = ['00000101010101010101', '00101010101101110111', '00100010010000001001']

muestra = input("Ingrese secuencia: ")

parentezco = []
maximo = 0
culpable = ""

for i in range(len(cromosomas)):
    contador = 0
    for j in range(20):
        if cromosomas[i][j] == muestra[j]:
            contador += 1
    porcentaje = contador * 100 / 20
    parentezco.append(porcentaje)
    if porcentaje > maximo:
        maximo = porcentaje
        culpable = nombres[i]

print("El sospechoso es", culpable, "con un porcentaje de parentesco del", maximo, "%")
```
